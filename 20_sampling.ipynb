## Code to Create a 20% Stratified Split

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your preprocessed files
train = pd.read_csv("train_preprocessed.csv")
test = pd.read_csv("test_preprocessed.csv")

target_col = "RiskFlag"

# Split into X and y
X_train = train.drop(columns=[target_col])
y_train = train[target_col]

X_test = test.drop(columns=[target_col])
y_test = test[target_col]

# ---- 20% STRATIFIED TRAIN ----
X_train_small, _, y_train_small, _ = train_test_split(
    X_train, y_train,
    test_size=0.80,
    stratify=y_train,
    random_state=42
)

# ---- 20% STRATIFIED TEST ----
X_test_small, _, y_test_small, _ = train_test_split(
    X_test, y_test,
    test_size=0.80,
    stratify=y_test,
    random_state=42
)

# Recombine
train_small = X_train_small.copy()
train_small[target_col] = y_train_small

test_small = X_test_small.copy()
test_small[target_col] = y_test_small

# ---- SAVE CSV FILES SO YOU CAN DOWNLOAD ----
train_small.to_csv("train_small_20.csv", index=False)
test_small.to_csv("test_small_20.csv", index=False)

print("Saved: train_small_20.csv")
print("Saved: test_small_20.csv")
```
